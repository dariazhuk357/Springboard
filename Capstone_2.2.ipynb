```python
# dropping unnecessary columns 
prefixes=['skills','abilities','interests','knowledge','acitivity','context']
df_list=[df_skills,df_abilities,df_interests,df_knowledge,df_activities,df_context]

def dropper(df,prefix): 
    irrelevant_columns=['N','Element ID','Scale ID','Recommend Suppress','Date','Domain Source']
    for column in df.columns: 
        if column in irrelevant_columns: 
            df.drop(columns=column,inplace=True)
        df['category']=prefix
```
