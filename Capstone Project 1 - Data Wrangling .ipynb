```python
import pandas as pd 
from IPython.display import display
import matplotlib.pyplot as plt
from datetime import datetime as dt 
from matplotlib.font_manager import FontProperties
import seaborn as sns 
%matplotlib inline 
```

```python
df=pd.read_csv('/Users/dasha/Downloads/Crime_Data_from_2010_to_Present.csv',parse_dates=['Date Occurred','Time Occurred'])
```

```python
pd.options.display.max_columns=None 
```

```python
df.info()
```

```python
df=df.drop(columns=['DR Number','Date Reported','Reporting District','MO Codes'])
df
```

```python
df['Time Occurred']=df['Time Occurred'].astype('str')
df['Time Occurred']=pd.to_datetime(df['Time Occurred'],format='%H%M').dt.time
df
```

```python
# Crime levels accross years, months, weeks organized by area name...
```

```python
timed_data=df.loc[:,['Date Occurred','Area Name','Crime Code Description']]
timed_data['Year']=timed_data['Date Occurred'].dt.year
```

```python
#Yearly data....
```

```python
yearly_data=timed_data.drop(columns='Date Occurred')
yearly_data=yearly_data[yearly_data['Year']!=2018]
yearly_data=yearly_data.groupby(['Year','Area Name']).count()
yearly_data.head() 
```

```python
yearly_data['Crime Count']=yearly_data['Crime Code Description']
yearly_data=yearly_data.drop(columns='Crime Code Description')
```

```python
yearly_data=yearly_data.unstack(level='Area Name')
yearly_data
```

```python
plot=yearly_data['Crime Count'].plot(by='Area Name',style=['--','-.','--','-',':','-.','--','-',':','-.','--','-',':','-.','--','-',':','-.','--','-',':','-.','--','-'],figsize=(12,12))
plot.legend(loc='center left', bbox_to_anchor=(1, 0.5),prop={'size': 8})
plt.ylabel('Crime Count')
plt.title('LA Crime Rate by District')
```
